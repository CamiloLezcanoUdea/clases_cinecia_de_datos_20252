```python
# Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# 1. Grafico linea basico

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Grafico lineal")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
# 2. Cambiar color y estilo

plt.plot(x, y, color= 'red', linestyle= '-.', linewidth= 2)
plt.show()
```

```python
# 3. mostrar ecuaciones si ya las conozco

x1 = np.linspace(-5, 5, 100)
y1 = 2*x1 + 1

plt.plot(x1, y1)

plt.text(0, 0, r"$y = 2x + 1$", fontsize= 12)
plt.text(0, -2, r"$R^2 = 1$", fontsize = 12)

plt.title("Ecuacion conocida")
plt.grid(True)
plt.show()
```

```python
# 4. mostrar ecuacion y R^2 solo con los datos (polinomio grado 1)


x2 = np.array([1, 2, 3, 4, 5])
y2 = np.array([2.2, 4.1, 6.05, 8.1, 10.2])

# regresion lineal

m, b = np.polyfit(x2, y2, 1)
y2_pred = m*x2 + b

# calcular R^2

SS_res = np.sum((y2 - y2_pred)**2)
SS_tot = np.sum((y2 - np.mean(y2))**2)
r2 = 1 - ((SS_res / SS_tot))

plt.scatter(x2, y2)
plt.plot(x2, y2_pred, color= 'red')

eq = rf"$y = {m:.2f}x + {b:.2f}$"
r2_text = rf"$R^2 = {r2:.4f}$"
plt.text(1, 10, eq, fontsize = 12)
plt.text(1, 8, r2_text, fontsize = 12)

plt.title("Ecuacion conocida")
plt.xlabel("tiempo")
plt.ylabel("concentracin")
plt.grid(True)
plt.show()
```

```python
# Seaborn

import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")

# 1. Dispesion simple

sns.scatterplot(data = df, x = "total_bill", y = "tip")
plt.show()
```

```python
# 2. dispersion con color

sns.scatterplot(data = df, x = "total_bill", y = "tip", hue = "sex")
plt.show
```

```python
# 4. barplot promedio

sns.barplot(data = df, x = "day", y = "total_bill")
plt.show()
```

```python
#5. countplot

sns.countplot(data = df, x = "day")
plt.show()
```
