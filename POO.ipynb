POO - Programación orientada a objetos

```python
# @title
# Definicion de clase

class Perro:
  # Constructor: se ejecuta cunado se crea la instacia de clase
  def __init__(self, nombre, raza, edad):
    self.nombre = nombre
    self.raza = raza
    self.edad = edad

  # Metodo de la clase perro
  def ladrar(self):
      print("Guau")

  def mostrar_informacion(self):
      print(f"Nombre: {self.nombre}, Raza: {self.raza}, Edad: {self.edad}")


firulais = Perro("Firulais", "Labrador", 4)
toby = Perro("Toby", "Bulldog", 2)

firulais.mostrar_informacion()
firulais.ladrar()
toby.mostrar_informacion()
toby.ladrar()
```

```python
# 1. Abstraccion

class Vehiculo:
  def __init__(self, marca, modelo, año):
    self.marca = marca
    self.modelo = modelo
    self.año = año

  def velocidad(self, vel):
    print("El vehiculo tiene una velocidad maxima de", vel)

carro1 = Vehiculo("Toyota", "Corolla", 2022)
carro1.velocidad(200)
```

```python
# 2. Encapsulamiento

class Gato:
  def __init__(self, nombre, raza):
    self.__nombre = nombre
    self.__raza = raza

  def get_nombre(self):
    return self.__nombre

  def set_nombre(self, nuevo_nombre):
    if len(nuevo_nombre) > 0:
      self.__nombre = nuevo_nombre
    else:
      print("El nombre no puede estar vacio")

  def maullar(self):
    print(f"{self.__nombre} dice ¡Miau!")

gato1 = Gato("Whiskers", "Siamés")
gato1.maullar()
gato1.set_nombre("Tom")
gato1.maullar()
```

```python
# 3. Herencia

class Animal:
  def __init__(self, nombre, edad, especie):
    self.nombre = nombre
    self.edad = edad
    self.especie = especie

  def saltar(self):
    print(f"{self.nombre} esta saltando")

  def comunicarse(self, mensaje):
    print(f"{self.nombre} dice: {mensaje}")

# Nueva clase hija de Animal
class Hamster(Animal):
  def __init__(self, nombre, edad, especie, dueno):
    super().__init__(nombre, edad, "Hamster")
    self.dueno = dueno

  def comer_semillas(self):
    print(f"{self.nombre} esta comiendo semillas")

hamster1 = Hamster("Hammy", 2, "Hamster", "Juan")
hamster1.saltar()
hamster1.comer_semillas()
```

```python
# 4. Polimorfismo

class vaca(Animal):
  def __init__(self, nombre, edad, especie):
    super().__init__(nombre, edad, especie)

class pato(Animal):
  def __init__(self, nombre, edad, especie):
    super().__init__(nombre, edad, especie)

vaca1 = vaca("Lola", 3, "Vaca")
pato1 = pato("Piolin", 1, "Pato")

vaca1.comunicarse("Muuu")
pato1.comunicarse("Quack")
```

```python
# Principios SOLID

# S: Single Responsability
# O: Open/Closed
# L: Liskov Substitution
# I: Interface Segregation
# D: Dependency Inversion
```

```python
# 1. S: Single Responsability

class Impresora:
  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo

  def imprimir(self, texto):
    print(f"{texto}")
```

```python
# 2. O: Open/Closed

class vehiculo:
  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo

  def velocidad(self, vel):
    print(f"El vehiculo tiene una velocidad maxima de {vel}")

# Extension

class bicicleta(vehiculo):
  def __init__(self, marca, modelo):
    super().__init__(marca, modelo)

  def velocidad(self, vel):
    print(f"La bicicleta tiene una velocidad maxima de {vel}")
```

```python
# 3. Liskov Substitution

class pollito(Animal):
  def __init__(self, nombre, edad, especie):
    super().__init__(nombre, edad, especie)

pollito1 = pollito("Piolin", 3, "Pollito")
pollito1.comunicarse("Pio Pio")
```
